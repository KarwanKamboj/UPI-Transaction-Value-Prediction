```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score


df = pd.read_csv(r"D:\Predictive_project\rbi_daily_digital_payments_dataset (1).csv")
```

```python
print("Shape :",df.shape)
print(df.describe())
```

```python
df.isnull().sum()
```

```python
# df.drop(columns=['Unnamed: 0'], inplace=True)
print(df.shape) # index column has been removed
```

```python

df['date'] = pd.to_datetime(df['date'], errors='coerce', dayfirst=True)

df.sort_values('date', inplace=True)

df.fillna(0, inplace=True)
```

```python
df.dtypes
```

```python
df.head() # To check that NaN values repalced with 0 or not
```

```python
# Boxplot to detect outliers

plt.figure(figsize=(7,5))
plt.boxplot(df['upi_val'])
plt.xlabel("UPI Transaction Value")
plt.title("Box Plot of UPI Transaction Value")
plt.show()
```

```python
X = df.drop(columns=['upi_val', 'date'])   # Independent variables
y = df['upi_val']                          # Dependent variable
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# STEP 11: Model Evaluation
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# STEP 12: Print Results
print("Model Performance:")
print("MAE  :", mae)
print("MSE  :", mse)
print("RMSE :", rmse)
print("R2 Score :", r2)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X_scaled, y, cv=5, scoring='r2')
print("Average R2 Score:", scores.mean())
```

```python
# STEP 13: Compare Actual vs Predicted values

comparison = pd.DataFrame({
    'Actual_UPI_Value': y_test.values[:10],
    'Predicted_UPI_Value': y_pred[:10]
})

print("\nActual vs Predicted UPI Values (Sample):")
print(comparison)
```

```python
# STEP 14: Predict UPI value for a new sample input

sample_input = X_test[0].reshape(1, -1)
predicted_value = model.predict(sample_input)

print("\nPredicted UPI value for new input:", predicted_value[0])
```

```python
# Actual vs Predicted Graph
plt.figure(figsize=(7,5))
plt.scatter(y_test, y_pred, color='green', alpha=0.6)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()],
         color='red', linestyle='--')

plt.xlabel("Actual UPI Value")
plt.ylabel("Predicted UPI Value")
plt.title("Actual vs Predicted UPI Transaction Value")
plt.show()
```

```python
# Residual Error Plot

errors = y_test - y_pred

plt.figure(figsize=(7,5))
plt.scatter(y_pred, errors, color='purple', alpha=0.6)
plt.axhline(y=0, color='black', linestyle='--')

plt.xlabel("Predicted UPI Value")
plt.ylabel("Prediction Error")
plt.title("Residual Error Plot")
plt.show()
```
